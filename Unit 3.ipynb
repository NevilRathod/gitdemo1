# Text Annotation

```python
#draw plot as fig #if we want to share it with external public we can do that using this function ,as it can give us pdf , jpg format

fig,ax=plt.subplots()
x=np.linspace(0,20,100) #(start,end,number of points)
y=np.cos(x)
ax.plot(x,y)
ax.axis('equal') #it will equally proportionate on both axis
ax.annotate("local maximum",xy=(2*np.pi,1),xytext=(10,4),arrowprops=dict(facecolor='black',shrink=5))
ax.annotate("local minimum",xy=(5*np.pi,-1),xytext=(10,-4),arrowprops=dict(facecolor='black',shrink=5))
plt.show()
```

# 3D Plotting

```python
#import mplot3d toolkit,numpy,matplotlib

from mpl_toolkits import mplot3d  #to plot 3d plots
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
#3d projection

ax = plt.axes(projection='3d')
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'red')
ax.set_title('3D line plot')
plt.show()
```

# Pie Chart

```python
#Creating a data set

Country = ['India','China','Pakistan','USA','Russia']
Population= [131,135,95,40,50]
explode =(0,0.1,0,0,0)          #to highlight a slice we use explode...!!
                                #autopct=autopercentage
    
plt.pie(Population,labels=Country,explode=explode,autopct='%1.1f%%')
fig= plt.figure(figsize=(15,12))
plt.show()
```

```python
Country = ['India','China','Pakistan','USA','Russia']
Population= [131,135,95,40,50]
explode =(0,0,0,0.1,0)
plt.pie(Population,labels=Country,explode=explode)
fig= plt.figure(figsize=(15,12))
plt.show()
```

```python
Country = ['India','China','Pakistan','USA','Russia']
Population= [131,135,95,40,50]
explode =(0,0.1,0,0,0)
plt.pie(Population,labels=Country,explode=explode,shadow=True)
fig= plt.figure(figsize=(15,12))
plt.show()
```

```python
Country = ['India','China','Pakistan','USA','Russia']
Population= [131,135,95,40,50]
explode =(0,0.1,0,0,0)
plt.pie(Population,labels=Country,explode=explode,shadow=True,autopct='%1.1f%%')
fig= plt.figure(figsize=(15,12))
plt.show()
```
